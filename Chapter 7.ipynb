# Example of Latex in plot

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np
```

```python
def f(x):
    return 0.5 * np.exp(x) + 1
```

```python
a, b = 0.5, 1.5
x = np.linspace(0, 2)
y = f(x)
Ix = np.linspace(a, b)
Iy = f(Ix)
vertices = [(a, 0), *zip(Ix, Iy), (b, 0)]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y)
plt.ylim(bottom=0)
poly = Polygon(vertices, facecolor="0.7", edgecolor="0.5")
ax.add_patch(poly)
plt.text(0.5 * (a + b), 1, "$\int^b_a f(x)dx$", horizontalalignment="center", fontsize=20)
plt.show()
```
